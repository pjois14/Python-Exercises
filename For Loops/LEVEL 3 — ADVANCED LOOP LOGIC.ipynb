1.	Check if a given number is prime (using a for loop).

```python
numbers = [1, 3, 7, 2, 8, 4, 6, 5, 5] 
prime_numbers=[]
non_prime_numbers=[]
for n in numbers:
    if n<=3:
        prime_numbers.append(n)
    else:
        for i in range(2,int(n**0.5)+1):
            if n%i == 0:
                non_prime_numbers.append(n)
                non_prime_numbers.sort()
            else:
                prime_numbers.append(n)
                prime_numbers.sort()
print(f" Prime numbers from the provided list are : {prime_numbers}")
print(f" Non Prime Numbers from the provided list are : {non_prime_numbers}")
    
```

22.	Find all prime numbers between 1 and 100.

```python
prime_numbers_ranged = []
non_prime_numbers_ranged = []
for n in range(1,101):
    if n<=3:
        prime_numbers_ranged.append(n)
    else:
        for i in range(2,int(n**0.5)+1):
            is_prime = True
            if n%i == 0:
                non_prime_numbers_ranged.append(n)
                is_prime = False
                break;
        if is_prime == True:
            prime_numbers_ranged.append(n)
                    
print(f" Prime numbers from the provided list are : {prime_numbers_ranged}")
print(f" Non Prime Numbers from the provided list are : {non_prime_numbers_ranged}")
```

25.	Compute the sum of the squares of all elements in a list.

```python
numbers = [1, 3, 7, 2, 8, 4, 6, 5] 
squared_value = 0
total_value = 0 
for n in numbers:
    squared_value = n * n
    print(f"The square of number n is : {squared_value}")
    total_value = total_value + squared_value

print(f"The sum of the squares of all elements in a list is {total_value}")
```
