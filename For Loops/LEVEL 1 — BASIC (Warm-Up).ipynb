7.	Calculate the average (mean) of numbers in a list.

```python
numbers = [5, 10, 35, -20, 25]
```

```python
sum = 0;
counter = 0;
average = 0;
for x in numbers:
    sum = sum + x
    counter = counter+1
print (f"Sum of the numbers is {sum}")
print (f"Total no of values are {counter}")
average = sum / counter;
print (f"Average of the provided list is {average}")
```

8.	Find both the maximum and minimum in a single loop (without built-in max() or min()).

```python
min_number=numbers[0]
max_number=numbers[0]
for x in numbers:
    if x > max_number:
        max_number = x
    if x < min_number:
        min_number = x
print(f" minimum number from the list is {min_number}")
print(f" Maximum number from the list is {max_number}")
```

9.	Multiply all numbers in a list (compute the product).

```python
product = 1
for x in numbers:
    product = product * x
print(f"Product of all the values present in the list is : {product}")
```

```python
if len(numbers) == 0:
    print(f"the provided list does not have any Values !!! ")
else:
    product = 1
    for x in numbers:
        product = product * x
    print(f"Product of all the values present in the list is : {product}")
    
```
